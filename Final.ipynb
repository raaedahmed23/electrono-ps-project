```python

def beautify(df):
    df.signalname = [s[:-16] if len(s)>24 else s for s in df.signalname]
    df.updatedate = [str(d)[11:-3] for d in df.updatedate]
    df.enddate = [str(d)[11:-3] for d in df.enddate]
    df._id = [str(d)[10:-2] for d in df._id]

    time_up = [val.replace('-', ':') for val in df.updatedate.values]
    time_up = [val.replace('T', ':') for val in time_up]
    time_up = [val.replace('.', ':') for val in time_up]
    df.updatedate = time_up

    time_up = [val.replace('-', ':') for val in df.enddate.values]
    time_up = [val.replace('T', ':') for val in time_up]
    time_up = [val.replace('.', ':') for val in time_up]
    df.enddate = time_up
    return df
df = beautify(df)
```

```python

def getValues(df, startTime, endTime, signalName):
    df_smaller = df.loc[df.signalname == signalName]
    val = np.array([])
    for row in df_smaller.itertuples():
        if((str(row[3]) >= startTime) and (str(row[4]) <= endTime)):
            num = int(row[5]/0.5)
            for j in range(num):
                val = np.append(val, row[7])
    return val
```

```python
def make2Darray(df, signalName):
    df_time = df[df.signalname=='MainProgram_path1']
    arr2D = []
    for row in df_time.itertuples():
        if(row[5]<400):
            start_time = row[3]
            end_time = row[4]
            arr = getValues(df, start_time, end_time, signalName)
            arr2D.append(arr)
    return arr2D
speed = make2Darray(df, 'SpindleLoad_0_path1')
speed
```

```python
## making a 244x697 2d array with unavaiable values as NaN
def resizeArray(speed):
    speed_arr = []
    for i in range(len(speed)):
        val = []
        for j in range(697):
            if j<len(speed[i]):
                val.append(speed[i][j])
            else:
                val.append(np.nan)
        speed_arr.append(val)
    speed_arr = np.array(speed_arr)
    return speed_arr
#speed_arr
        
```

```python
 ##finding the mean of each of the 697 columns
def meanColumns(speed_arr):
    mean = []
    for i in range(len(speed_arr.T)):
        mean.append(np.nanmean(speed_arr[:,i]))
    return mean
```

```python
 ##finding the std of each of the columns
def stdColumns(speed_arr):
    std = []
    for i in range(len(speed_arr.T)):
        std.append(np.nanstd(speed_arr[:,i]))
    return std
```

```python
def maxColumns(speed_arr):
    cmax = []
    for i in range(len(speed_arr.T)):
        cmax.append(np.nanmax(speed_arr[:,i]))
    return cmax
```

```python
def minColumns(speed_arr):
    cmin=[]
    for i in range(len(speed_arr.T)):
        cmin.append(np.nanmin(speed_arr[:,i]))
    return cmin
```
